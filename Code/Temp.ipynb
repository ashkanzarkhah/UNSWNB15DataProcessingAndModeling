```python
#Loading data
import pandas as pd

df_train = pd.read_csv("../Dataset/UNSW_NB15_training-set.csv")
df_test = pd.read_csv("../Dataset/UNSW_NB15_testing-set.csv")
df_all = pd.concat([df_train, df_test], ignore_index=True)

#Learning about the data
print(df_train)
print(df_test)
print(df_all)
```

```python
#Learning about categorical features
import numpy as np
columns_train = df_train.columns
for column in columns_train:
    if(type(df_train[column][0]) == type("")):
        print(column)
        unique_values_train = df_train[column].unique()
        unique_values_test = df_test[column].unique()
        print(np.sort(unique_values_train))
        print(np.sort(unique_values_test))
        print("DIFF TRAIN WITH TEST:")
        diff_arr1 = np.setdiff1d(unique_values_train, unique_values_test)
        print(diff_arr1)
        print("DIFF TEST WITH TRAIN:")
        diff_arr2 = np.setdiff1d(unique_values_test, unique_values_train)
        print(diff_arr2)
        print("######")
```

```python
#Checking if data has any missing values
print("Number of missing values in total:", df_all.isnull().sum().sum())
```

```python
#Now it's time to do preprocessings
#Fist we get rid of categorical features by assigning unique numbers to them.
columns_all = df_all.columns
for column in columns_all:
    if(type(df_all[column][0]) == type("")):
        df_all[column] = pd.factorize(df_all[column])[0]
        unique_values_all = df_all[column].unique()
        print(np.sort(unique_values_all))
        print("####")
```

```python
#Second we see if features need to be scaled
for column in columns_all:
    print(column, "Max:", df_all[column].max(), "Min:", df_all[column].min())
```

```python
#As we can see, they vary a lot, so I decide to do a min-max scale on it and make their intervals [0, 1]
#except our lable and first feature witch is an id and we will remove it

df_all_scaled = df_all.apply(lambda x: (x - x.min()) / (x.max() - x.min()))
df_all_scaled[columns_all[0]] = df_all[columns_all[0]]
df_all_scaled[columns_all[-1]] = df_all[columns_all[-1]]
```

```python
#Removing id column
df_all_scaled = df_all_scaled.drop(columns_all[0], axis=1)
print(df_all_scaled)
```

```python
#Now we check our data's balancement

cnt = 10 * [0]
for lable in df_all_scaled[columns_all[-1]]:
    cnt[lable] += 1
    
for i in range(10):
    print("#lable", i, " :", cnt[i])
```

```python
#As we can see lables 1, 2, 4, and 8 are very imbalance and we decide to cut them off from the dataset
#But first we need to devide our data into df_train and df_test

df_train_all = df_all_scaled.iloc[:len(df_train)]
df_test_all = df_all_scaled.iloc[len(df_train):]
```

```python
df_train = df_train_all[df_train_all[columns_all[-1]] != 1]
df_train = df_train[df_train[columns_all[-1]]  != 2]
df_train = df_train[df_train[columns_all[-1]]  != 4]
df_train = df_train[df_train[columns_all[-1]]  != 8]

cnt = 10 * [0]
for lable in df_train[columns_all[-1]]:
    cnt[lable] += 1
    
for i in range(10):
    print("#lable", i, " :", cnt[i])
    
print("######")    
df_test = df_test_all[df_test_all[columns_all[-1]] != 1]
df_test = df_test[df_test[columns_all[-1]]  != 2]
df_test = df_test[df_test[columns_all[-1]]  != 4]
df_test = df_test[df_test[columns_all[-1]]  != 8]

cnt = 10 * [0]
for lable in df_test[columns_all[-1]]:
    cnt[lable] += 1

for i in range(10):
    print("#lable", i, " :", cnt[i])
    
df_train = df_train.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)

print(df_train)
print(df_test)
```

```python
#Now that our preprocessing is over, it's time to devide our train and test data
#into X_train, y_train, X_test, and y_test

X_train = df_train.filter(items = df_train.columns[:-1])
y_train = df_train.filter(items = df_train.columns[-1:])

X_test = df_test.filter(items = df_test.columns[:-1])
y_test = df_test.filter(items = df_test.columns[-1:])
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.datasets import load_iris

# Create a random forest classifier as the underlying model
rf = RandomForestClassifier(random_state=42)

# Create the SelectFromModel object, specifying the random forest model
selector = SelectFromModel(estimator=rf)

# Fit the selector to the data
selector.fit(X_train, y_train)

# Transform the data to select the important features
X_selected = selector.transform(X_train)

# Get the support mask indicating selected features
feature_mask = selector.get_support()

# Get the indices of the selected features
selected_feature_indices = [i for i, mask in enumerate(feature_mask) if mask]

# Print the selected feature indices
print("Selected feature indices:", selected_feature_indices)
```

```python
def array_union(arr1, arr2):
    # Concatenate the two arrays
    combined_array = arr1 + arr2

    # Create a new empty array for the union elements
    union_array = []

    # Iterate through each element in the combined array
    for element in combined_array:
        # Check if the element is not already in the union array
        if element not in union_array:
            # Add the element to the union array
            union_array.append(element)

    return union_array
#selected_feature_indices = [6, 7, 8, 9, 12, 26, 31, 33, 34, 40]

selected_feature_indices = array_union(selected_feature_indices, selected_feature_indices2)
print(selected_feature_indices)
                           
X_train_selected = X_train.iloc[:, selected_feature_indices]
X_test_selected = X_test.iloc[:, selected_feature_indices]
print(X_train_selected)

# Create a new Random Forest classifier with weighted and selected features
weighted_random_forest = RandomForestClassifier(criterion = 'entropy',
                               n_estimators = 100,
                               class_weight='balanced',
                               n_jobs = -1)

# Fit the new Random Forest classifier on the selected features
weighted_random_forest.fit(X_train_selected, np.array([x[0] for x in np.array(y_train)]))

# Make predictions on the test data using the selected features
y_pred = weighted_random_forest.predict(X_test_selected)
y_pred_train = weighted_random_forest.predict(X_train_selected)
```

```python
#Getting model results
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

accuracy = accuracy_score(y_test, y_pred)
accuracy_train = accuracy_score(y_train, y_pred_train)

precision = precision_score(y_test, y_pred, average = "macro")
precision_train = precision_score(y_train, y_pred_train, average = "macro")

recall = recall_score(y_test, y_pred, average = "macro")
recall_train = recall_score(y_train, y_pred_train, average = "macro")

f1 = f1_score(y_test, y_pred, average = "macro")
f1_train = f1_score(y_train, y_pred_train, average = "macro")

confusion = confusion_matrix(y_test, y_pred)
confusion_train = confusion_matrix(y_train, y_pred_train)

print("accuracy", accuracy)
print("accuracy_train", accuracy_train)
print("precision",precision)
print("precision_train",precision_train)
print("recall", recall)
print("recall_train", recall_train)
print("f1", f1)
print("f1_train", f1_train)
print("confusion", confusion)
print("confision_train", confusion_train)
```
