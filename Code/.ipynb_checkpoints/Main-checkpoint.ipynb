```python
#Loading data
import pandas as pd

df_train = pd.read_csv("../Dataset/UNSW_NB15_training-set.csv")
df_test = pd.read_csv("../Dataset/UNSW_NB15_testing-set.csv")
df_all = pd.concat([df_train, df_test], ignore_index=True)

#Learning about the data
print(df_train)
print(df_test)
print(df_all)
```

```python
#Learning about categorical features
import numpy as np
columns_train = df_train.columns
for column in columns_train:
    if(type(df_train[column][0]) == type("")):
        print(column)
        unique_values_train = df_train[column].unique()
        unique_values_test = df_test[column].unique()
        print("DIFF TRAIN WITH TEST:")
        diff_arr1 = np.setdiff1d(unique_values_train, unique_values_test)
        print(diff_arr1)
        print("DIFF TEST WITH TRAIN:")
        diff_arr2 = np.setdiff1d(unique_values_test, unique_values_train)
        print(diff_arr2)
        print("######")
```

```python
#Checking if data has any missing values
print("Number of missing values in total:", df_all.isnull().sum().sum())
```

```python
#Now it's time to do preprocessings
#Fist we get rid of categorical features by assigning unique numbers to them.
columns_all = df_all.columns
for column in columns_all:
    if(type(df_all[column][0]) == type("")):
        df_all[column] = pd.factorize(df_all[column])[0]
        unique_values_all = df_all[column].unique()
        print(np.sort(unique_values_all))
        print("####")
```

```python
#Second we see if features need to be scaled
for column in columns_all:
    print(column, "Max:", df_all[column].max(), "Min:", df_all[column].min())
```

```python
#As we can see, they vary a lot, so I decide to do a min-max scale on it and make their intervals [0, 1]
#except our lable and first feature witch is an id and we will remove it

df_all_scaled = df_all.apply(lambda x: (x - x.min()) / (x.max() - x.min()))
df_all_scaled[columns_all[0]] = df_all[columns_all[0]]
df_all_scaled[columns_all[-1]] = df_all[columns_all[-1]]
```

```python
#Removing id column
df_all_scaled = df_all_scaled.drop(columns_all[0], axis=1)
print(df_all_scaled)
```

```python
#Now we check our data's balancement

cnt = 10 * [0]
for lable in df_all_scaled[columns_all[-1]]:
    cnt[lable] += 1
    
for i in range(10):
    print("#lable", i, " :", cnt[i])
```

```python
#As we can see lables 1, 2, 4, and 8 are very imbalance and we decide to cut them off from the dataset
#But first we need to devide our data into df_train and df_test

df_train_all = df_all_scaled.iloc[:len(df_train)]
df_test_all = df_all_scaled.iloc[len(df_train):]

print(df_train_all)

df_train = df_train_all[df_train_all[columns_all[-1]] != 1]
df_train = df_train[df_train[columns_all[-1]]  != 2]
df_train = df_train[df_train[columns_all[-1]]  != 4]
df_train = df_train[df_train[columns_all[-1]]  != 8]

cnt = 10 * [0]
for lable in df_train[columns_all[-1]]:
    cnt[lable] += 1
    
for i in range(10):
    print("#lable", i, " :", cnt[i])
    
print("######")    
df_test = df_test_all[df_test_all[columns_all[-1]] != 1]
df_test = df_test[df_test[columns_all[-1]]  != 2]
df_test = df_test[df_test[columns_all[-1]]  != 4]
df_test = df_test[df_test[columns_all[-1]]  != 8]

cnt = 10 * [0]
for lable in df_test[columns_all[-1]]:
    cnt[lable] += 1

for i in range(10):
    print("#lable", i, " :", cnt[i])
    
df_train = df_train.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)

print(df_train)
print(df_test)
        
```

```python
#Now that our preprocessing is over, it's time to devide our train and test data
#into X_train, y_train, X_test, and y_test

X_train = df_train.filter(items = df_train.columns[:-1])
y_train = df_train.filter(items = df_train.columns[-1:])

X_test = df_test.filter(items = df_test.columns[:-1])
y_test = df_test.filter(items = df_test.columns[-1:])
```

```python
#Now that we have our data, it's time to do some feature reduction and make
#some instances of our train data and build different models on them
#First we start with autoencoders and we use 2 layers

import keras
from keras import layers

#Defining input shape
input_img = keras.Input(shape=(42,))

#We use 2 layers for encoding and then 2 layers for decoding
encoded = layers.Dense(30, activation='relu')(input_img)
encoded = layers.Dense(21, activation='relu')(encoded)

decoded = layers.Dense(30, activation='relu')(encoded)
decoded = layers.Dense(42, activation='sigmoid')(decoded)

#Defining model's input and output shape
autoencoder = keras.Model(input_img, decoded)

#Fixing optimizer and loss function
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')

#Fitting our encoder base on Train data and loss value for both of our train and test data
autoencoder.fit(X_train, X_train,
                epochs=100,
                batch_size=256,
                shuffle=True,
                validation_data=(X_test, X_test))
```

```python
#Defining encoder and decoder base on learned encoding layers
encoder = keras.Model(input_img, encoded)
decoder = keras.Model(encoded, decoded)

#Encoding train and test data
encoded_train_imgs = encoder.predict(X_train)
encoded_test_imgs = encoder.predict(X_test)

#Decoding train and test data
decoded_train_imgs = decoder.predict(encoded_train_imgs)
decoded_test_imgs = decoder.predict(encoded_test_imgs)
```

```python
#Another way for feature reduction we are using is PCA
from sklearn.decomposition import PCA
def library_PCA(X, n):
    #Creating n*d matrix
    tmpX = X.copy()
    #Subtracting mean
    tmpX = tmpX - np.mean(tmpX, axis = 0)
    #Builiding pca
    pca = PCA(n_components=n)
    #Finding Components
    principalComponents = pca.fit_transform(tmpX)
    return principalComponents, pca

lib_transformed_30d_X_train, lib_fitted_30d_pca = library_PCA(X_train, 30)
lib_transformed_30d_X_test = lib_fitted_30d_pca.transform(X_test)

lib_transformed_20d_X_train, lib_fitted_20d_pca = library_PCA(X_train, 20)
lib_transformed_20d_X_test = lib_fitted_20d_pca.transform(X_test)
```

```python
#Another way for feature reduction we are using a supervised feature reduction
#in witch we use random forests feature importance for our less common lables
#and rfe for all of our data
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, random_state=42)

rfe = RFE(estimator=model, n_features_to_select=10)

rfe.fit(X_train, np.array([x[0] for x in np.array(y_train)]))

selected_feature_indices = rfe.get_support(indices=True)

print(selected_feature_indices)
```

```python
X_train_selected = df_train_all[df_train_all[columns_all[-1]] != 0]
X_train_selected = X_train_selected[X_train_selected[columns_all[-1]]  != 9]
X_train_selected = X_train_selected.filter(items = df_train.columns[:-1])

y_train_selected = df_train_all[df_train_all[columns_all[-1]] != 0]
y_train_selected = y_train_selected[y_train_selected[columns_all[-1]]  != 9]
y_train_selected = y_train_selected.filter(items = df_train.columns[-1:])

random_forest = RandomForestClassifier(n_estimators=100, random_state=42)

random_forest.fit(X_train_selected, np.array([x[0] for x in np.array(y_train_selected)]))

feature_importances = random_forest.feature_importances_

sorted_feature_indices = sorted(range(len(feature_importances)), key=lambda k: feature_importances[k], reverse=True)

selected_feature_indices2 = sorted_feature_indices[:10]

print("Selected Feature Indices:", selected_feature_indices2)
```

```python
def union(arr1, arr2):
    combined = arr1 + arr2
    union = []
    for element in combined:
        if element not in union:
            union.append(element)

    return union

selected_feature_indices = union(selected_feature_indices, selected_feature_indices2)
print(selected_feature_indices)
                           
X_train_selected = X_train.iloc[:, selected_feature_indices]
X_test_selected = X_test.iloc[:, selected_feature_indices]
```

```python
#Now that we have reducted our features, it's time to build some models on them
all_data_names = ["Main", "Encoded", "PCA 30d", "PCA 20d", "Selected"]
all_trains = [X_train, pd.DataFrame(encoded_train_imgs),
              pd.DataFrame(lib_transformed_30d_X_train),
              pd.DataFrame(lib_transformed_20d_X_train),
              X_train_selected
             ]

#First we start with SVM and we use 5-fold cross validation to find best parameters
from sklearn import svm
from sklearn import metrics
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.model_selection import cross_validate
from sklearn.model_selection import train_test_split


Kernels = ['linear', 'poly', 'rbf']
Degrees = [3, 4]

best_data = ''
best_Kernel = ''
best_Degree = 0
best_score = 0

best_data_f1 = ''
best_Kernel_f1 = ''
best_Degree_f1 = 0
best_score_f1 = 0

for i in range(5):
    _, data_x, _, data_y = train_test_split(all_trains[i], y_train, test_size=0.11, stratify = y_train)
    data_x = data_x.reset_index(drop=True)
    data_y = data_y.reset_index(drop=True)
    
    #data = pd.concat([all_trains[i], y_train], axis=1).sample(n=20000, random_state=42)
    #data_x = data.filter(items = data.columns[:-1])
    #data_y = data.filter(items = data.columns[-1:])
    
    for Kernel in Kernels:
        for Degree in Degrees:
            if(Kernel != 'poly' and Degree == 4):
                continue
            my_pipeline = Pipeline(steps=[('preprocessor', SimpleImputer()),
                                                  ('model', svm.SVC(kernel = Kernel,
                                                            degree = Degree))
                                                 ])
            
            print("Data: ", all_data_names[i],
                 "| Kernel: ", Kernel,
                 "| Degree:", Degree)
            
            scores = cross_validate(my_pipeline, data_x, 
                            np.array([x[0] for x in np.array(data_y)]),
                            cv=5,
                            scoring=['accuracy', 'precision_macro', 'recall_macro', 'f1_macro'],
                            return_train_score=True, return_estimator=True)
            
            accurancy_scores = scores['test_accuracy']
            precision_scores = scores['test_precision_macro']
            recall_scores = scores['test_recall_macro']
            f1_scores = scores['test_f1_macro']
            
            print("accurancy_scores:\n", accurancy_scores)
            print("precision_scores:\n", precision_scores)
            print("recall_scores:\n", recall_scores)
            print("f1_scores:\n", f1_scores)
            score = np.mean(accurancy_scores)
            print("Average accurancy score:", score)
            
            f1_score = np.mean(f1_scores)
            print("Average f1 score:", f1_score)
            
            if(score > best_score):
                best_Kernel = Kernel
                best_Degree = Degree
                best_data = all_data_names[i]
                best_score = score
            
            if(f1_score > best_score_f1):
                best_Kernel_f1 = Kernel
                best_Degree_f1 = Degree
                best_data_f1 = all_data_names[i]
                best_score_f1 = f1_score
    
```

```python
print("best_Kernel :", best_Kernel,
     "| best_Degree :", best_Degree,
     "| best_data :", best_data,
     "| best_score :", best_score)
print("best_Kernel_f1 :", best_Kernel_f1,
     "| best_Degree_f1 :", best_Degree_f1,
     "| best_data_f1 :", best_data_f1,
     "| best_score_f1 :", best_score_f1)
```

```python
#Now it's time to find best parameters for random forest
from sklearn.ensemble import RandomForestClassifier

criterions = ["gini", "entropy"]
n_estimators = [50, 100]
max_depths = [None, 10, 20]

best_data = ''
best_criterion = "gini"
best_n_estimators = 75
best_max_depth = None
best_score = 0

best_data_f1 = ''
best_criterion_f1 = "gini"
best_n_estimators_f1 = 75
best_max_depth_f1 = None
best_score_f1 = 0

for i in range(5):
    _, data_x, _, data_y = train_test_split(all_trains[i], y_train, test_size=0.5, stratify = y_train)
    data_x = data_x.reset_index(drop=True)
    data_y = data_y.reset_index(drop=True)
    
    #data = pd.concat([all_trains[i], y_train], axis=1).sample(n=20000, random_state=42)
    #data_x = data.filter(items = data.columns[:-1])
    #data_y = data.filter(items = data.columns[-1:])
    
    for current_criterion in criterions:
        for current_n_estimator in n_estimators:
            for current_max_depth in max_depths:
                my_pipeline = Pipeline(steps=[('preprocessor', SimpleImputer()),
                                          ('model', RandomForestClassifier(criterion = current_criterion,
                                                                           n_estimators = current_n_estimator,
                                                                           max_depth = current_max_depth,
                                                                           n_jobs = -1))
                                         ])

                print("Data: ", all_data_names[i],
                     "| Current Criterion: ", current_criterion,
                     "| Current number of estimators: ", current_n_estimator,
                     "| Max Depth: ", current_max_depth)

                scores = cross_validate(my_pipeline, data_x, 
                                np.array([x[0] for x in np.array(data_y)]),
                                cv=5,
                                scoring=['accuracy', 'precision_macro', 'recall_macro', 'f1_macro'],
                                return_train_score=True, return_estimator=True)

                accurancy_scores = scores['test_accuracy']
                precision_scores = scores['test_precision_macro']
                recall_scores = scores['test_recall_macro']
                f1_scores = scores['test_f1_macro']

                print("accurancy_scores:\n", accurancy_scores)
                print("precision_scores:\n", precision_scores)
                print("recall_scores:\n", recall_scores)
                print("f1_scores:\n", f1_scores)
                score = np.mean(accurancy_scores)
                print("Average accurancy score:", score)
                f1_score = np.mean(f1_scores)
                print("Average f1 score:", f1_score)

                if(score > best_score):
                    best_criterion = current_criterion
                    best_n_estimators = current_n_estimator
                    best_max_depth = current_max_depth
                    best_data = all_data_names[i]
                    best_score = score
                    
                if(f1_score > best_score_f1):
                    best_criterion_f1 = current_criterion
                    best_n_estimators_f1 = current_n_estimator
                    best_max_depth_f1 = current_max_depth
                    best_data_f1 = all_data_names[i]
                    best_score_f1 = f1_score
```

```python
#Now it's time to find best parameters for MLP
from sklearn.neural_network import MLPClassifier

Hidden_layer_sizes = [(100, 25)]
Learning_rates = ['constant', 'invscaling', 'adaptive']
Activations = ['tanh', 'relu']

best_data = ''
best_Hidden_layer_size = 0
best_Learning_rate = 0
best_Activation = 0
best_score = 0

best_data_f1 = ''
best_Hidden_layer_size_f1 = 0
best_Learning_rate_f1 = 0
best_Activation_f1 = 0
best_score_f1 = 0

for i in range(5):
    _, data_x, _, data_y = train_test_split(all_trains[i], y_train, test_size=0.2, stratify = y_train)
    data_x = data_x.reset_index(drop=True)
    data_y = data_y.reset_index(drop=True)
    
    #data = pd.concat([all_trains[i], y_train], axis=1).sample(n=20000, random_state=42)
    #data_x = data.filter(items = data.columns[:-1])
    #data_y = data.filter(items = data.columns[-1:])
    
    for Hidden_layer_size in Hidden_layer_sizes:
        for Learning_rate in Learning_rates:
            for Activation in Activations:
                my_pipeline = Pipeline(steps=[('preprocessor', SimpleImputer()),
                                  ('model', MLPClassifier(hidden_layer_sizes=Hidden_layer_size,
                                                                 learning_rate=Learning_rate,
                                                                 activation = Activation))
                                 ])

                print("Data: ", all_data_names[i],
                     "| Current Hidden_layer_size: ", Hidden_layer_size,
                     "| Current learning_rate: ", Learning_rate,
                     "| Current activation: ", Activation)

                scores = cross_validate(my_pipeline, data_x, 
                                np.array([x[0] for x in np.array(data_y)]),
                                cv=5,
                                scoring=['accuracy', 'precision_macro', 'recall_macro', 'f1_macro'],
                                return_train_score=True, return_estimator=True)

                accurancy_scores = scores['test_accuracy']
                precision_scores = scores['test_precision_macro']
                recall_scores = scores['test_recall_macro']
                f1_scores = scores['test_f1_macro']

                print("accurancy_scores:\n", accurancy_scores)
                print("precision_scores:\n", precision_scores)
                print("recall_scores:\n", recall_scores)
                print("f1_scores:\n", f1_scores)
                score = np.mean(accurancy_scores)
                print("Average accurancy score:", score)
                f1_score = np.mean(f1_scores)
                print("Average f1 score:", f1_score)

                if(score > best_score):
                    best_Hidden_layer_size = Hidden_layer_size
                    best_Learning_rate = Learning_rate
                    best_Activation = Activation
                    best_data = all_data_names[i]
                    best_score = score

                if(f1_score > best_score_f1):
                    best_Hidden_layer_size_f1 = Hidden_layer_size
                    best_Learning_rate_f1 = Learning_rate
                    best_Activation_f1 = Activation
                    best_data_f1 = all_data_names[i]
                    best_score_f1 = f1_score
```
